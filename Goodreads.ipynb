```python
import pandas as pd
from selenium import webdriver
```

```python
browser=webdriver.Chrome('chromedriver.exe')
```

```python
url='https://www.goodreads.com/choiceawards/best-books-2017'
browser.get(url)
```

# Book:
     category;Book name;URL to the book cover image;URL to the book description      

```python
Best_Book_2017=pd.DataFrame()
```

```python

parent=browser.find_elements_by_css_selector('div.category.clearFix')#get_attribute('category__copy')
for cat in parent:
    try:
        category=cat.find_element_by_css_selector('h4.category__copy').text
    except:
        category=None
    try:
        b_name=cat.find_element_by_css_selector('img.category__winnerImage').get_attribute('alt')
    except:
        b_name=None
    try:
        cover_image=cat.find_element_by_css_selector('img.category__winnerImage').get_attribute('src')
    except:
        cover_image=None
    try:
        desc_url=cat.find_element_by_css_selector('a').get_attribute('href')
    except:
        desc_url=None
    desc_book=pd.Series({
        'Category':category,
        'image_url':cover_image,
        'Book_description':desc_url,
        'book name':b_name
    })
    Best_Book_2017=Best_Book_2017.append(desc_book,ignore_index=True)
    #print(category,'book name:',b_name,'image_url',cover_image,desc_url)
Best_Book_2017


```

# fetching all the urls of each book

```python
parent=browser.find_elements_by_css_selector('div.category.clearFix')#get_attribute('category__copy')
book_desc_url=[]
for cat in parent:
    try:
        desc_url=cat.find_element_by_css_selector('a').get_attribute('href')
    except:
        desc_url=None
    book_desc_url.append(desc_url)
    print(book_desc_url)
```

```python
book_desc_url
```

# 2nd part

```python
Boook_DataBase=pd.DataFrame()
for i in book_desc_url:
    url=i
    browser.get(url)
    parent_desc=browser.find_element_by_css_selector('div.miniBook')#winner.stacked.pollVote.pollVote--borderBottom.box
    parent_desc1=browser.find_element_by_css_selector('div.pollContents')
    
    try:
        author=parent_desc.find_element_by_css_selector('span[itemprop="name"]').text#.get_attribute('name').text
    except:
        author=None
    try:
        description=parent_desc.find_element_by_css_selector('div.gcaBookDescription').text
    except:
        description
    try:
        votes=parent_desc.find_element_by_css_selector('span.greyText.gcaNumVotes').text
        votes1=int(votes.split(" ")[0].replace(",",""))
    except:
        votes1=None
    try:
        Total_votes=parent_desc1.find_element_by_css_selector('div.greyText.gcaNumVotes').text
        Total_votes1=int(Total_votes.split(" ")[0].replace(",",""))
    except:
        Total_votes=None
    percenrage=(votes1/Total_votes1)*100
    Book_data=pd.Series({'Author':author,'Description':description,
              'Votes':votes1,'Total Votes':Total_votes1,
               'Percentage of votes':percenrage
              })
    Boook_DataBase=Boook_DataBase.append(Book_data,ignore_index=True)
Boook_DataBase
    
    
    #print(author,description,'\n',votes1,'\n',Total_votes1)
    #print('\n percentage of votes',votes1%100*Total_votes1)
```

# Table1= Best_Book_2017 , Table2=Boook_DataBase


```python
Final_table=pd.concat([Best_Book_2017,Boook_DataBase],axis=1,ignore_index=False)
Final_table
```
