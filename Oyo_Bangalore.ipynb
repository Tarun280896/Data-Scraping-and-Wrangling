```python
import os
import time
import pandas as pd
from selenium import webdriver
os.getcwd()
```

```python
browser=webdriver.Chrome('chromedriver.exe')
```

```python
url='https://www.oyorooms.com/oyos-in-bangalore'
browser.get(url)
```

```python
parent=browser.find_elements_by_css_selector('div.mdl-grid.mdl-grid--no-spacing.width100')
len(parent)
```

```python
Oyo_bang=pd.DataFrame()
for i in range(1,26):
    url='https://www.oyorooms.com/oyos-in-bangalore?page=%d' % (i)
    browser.get(url)
    time.sleep(2)
    print(i)
    parent=browser.find_elements_by_css_selector('div.mdl-grid.mdl-grid--no-spacing.width100')
    
    for hotel in parent:

        try:
            hotel_name=hotel.find_element_by_css_selector('span.newHotelCard__hotelName').text
        except:hotel_name=None
        try:
            original_price=hotel.find_element_by_css_selector('span.newHotelCard__revisedPricing').text
        except:
            original_price=None
        try:
            discount_price=hotel.find_element_by_css_selector('div.newHotelCard__pricing.u-fw6').text
        except:
            discount_price=None
        try:
            discount_offer=hotel.find_element_by_css_selector('div.newHotelCard__discount').text
        except:
            discount_offer=None
        try:
            hotel_rating=hotel.find_element_by_css_selector('span.hotelRating__value').text
        except:
            hotel_rating=None
        try:
            hotel_rating_text=hotel.find_element_by_css_selector('span.hotelRating__subtext').text
        except:
            hotel_rating_text=None
        try:
            location=hotel.find_element_by_css_selector('span.newHotelCard__hotelAddress').text
        except:
            location=None
        try:
            rating_count=hotel.find_element_by_css_selector('div.hotelRating__count').text
        except:
            rating_count=None
        try:
            img_url=hotel.find_element_by_css_selector('img').get_attribute('src')
        except:
            img_url=None
        oyo_desc={
            'Hotel Name':hotel_name,
            'Location':location,
            'Original Price':original_price,
            'Discount Price':discount_price,
            'Discount Offer':discount_offer,
            'Hotel Rating':hotel_rating,
            'Rating Type':hotel_rating_text,
            'Total Ratings':rating_count,
            'Image':img_url
        }
        Oyo_bang=Oyo_bang.append(oyo_desc,ignore_index=True)
        

        #print(hotel_name,original_price,discount_price,discount_offer,hotel_rating,hotel_rating_text,location,rating_count,img_url)       
Oyo_bang
```
